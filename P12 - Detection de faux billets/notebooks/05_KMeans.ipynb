```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.metrics import (
    confusion_matrix,
    recall_score
)
```

```python
X_train = pd.read_pickle(r"..\data\processed\X_train_clean.pkl")
X_test = pd.read_pickle(r"..\data\processed\X_test_clean.pkl")
y_train = pd.read_pickle(r"..\data\processed\y_train.pkl")
y_test = pd.read_pickle(r"..\data\processed\y_test.pkl")
```

```python
model = KMeans(n_clusters=2, n_init=20, random_state=42)
model.fit(X_train)
```

```python
df_clusters = pd.DataFrame({
    'cluster': model.labels_,
    'is_fake': y_train.values
})
df_clusters.groupby('cluster').mean().reset_index()
```

```python
y_predict = pd.Series(model.predict(X_test)).map({0: True, 1: False})
```

```python
cm_pct = confusion_matrix(y_test, y_predict, normalize='true')
cm_count = confusion_matrix(y_test, y_predict)

counts = ["{0:0.0f}\n".format(value) for value in cm_count.flatten()]
percentages = ["{0:.1%}".format(value) for value in cm_pct.flatten()]
box_labels = [f"{v1}{v2}" for v1, v2 in zip(counts, percentages)]
box_labels = np.asarray(box_labels).reshape(cm_pct.shape[0], cm_pct.shape[1])

plt.figure(figsize=(8, 5), dpi=100)
plt.title("Matrice de confusion (K-means)")
sns.heatmap(cm_pct, vmin=0.0, vmax=1.0, cmap='Blues',
            annot=box_labels, fmt='',
            xticklabels=["Vrai", "Faux"], yticklabels=["Vrai", "Faux"])
plt.xlabel("Valeurs prédites")
plt.ylabel("Valeurs réelles")
plt.show()
```

```python
score = recall_score(y_test, y_predict)
print(f"recall_score : {100 * score:.2f}%")
```
